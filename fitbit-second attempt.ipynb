```python
def getnumber1(str):
    if(str=='no'):
        return 1
    if(str=='yes'):
        return 2
    else:
        return 3
```

```python
def getnumber2(str):
    if(str=='carlitos'):
        return 1
    if(str=='adelmo'):
        return 2
    if(str=='eurico'):
        return 3
    if(str=='pedro'):
        return 4
    if(str=='charles'):
        return 5
    if(str=='jeremy'):
        return 6
    if(str==''):
        return 7
    else:
        return 8
```

```python
#applying function getnumber1 on new_window column to convert string to int
test['newwindow']=test['new_window'].apply(getnumber1)
del test["new_window"]
```

```python
##applying function getnumber2 on user_name column to convert string to int
test['username']=test['user_name'].apply(getnumber2)
del test["user_name"]
```

```python
#copying dataset to A
A=test
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn import cross_validation
from sklearn import datasets
from sklearn import ensemble
from sklearn import tree
import pydotplus
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC 
from sklearn.metrics import classification_report
```

```python
#splitting training and testing data
x_train,x_test,y_train,y_test = cross_validation.train_test_split(A,B,test_size=0.01)
```

```python
test.isnull().sum()
```

```python
#fitting and predicting and calculating score using dtree
clf1= tree.DecisionTreeClassifier()
clf1.fit(x_train,y_train)
y_pred1=clf1.predict(x_test)
score1=clf1.score(x_test,y_test)
print('y_pred1 - ',y_pred1)
print('score1 - ',score1)
```

```python
#making dtree
dot_data=tree.export_graphviz(clf1,out_file=None,
                              #feature_names=iris.feature_names,
                              #class_names= iris.target_names,
                              filled=True,
                              rounded =True,special_characters=True)
graph= pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf("fitbit.pdf")
```

```python
#fitting and predicting and calculating score using random forest classification
clf2=ensemble.RandomForestClassifier()
clf2.fit(x_train,y_train)
score2=clf2.score(x_test,y_test)
y_pred2=clf2.predict(x_test)
print('y_pred2 - ',y_pred2)
print('score2 - ',score2)
```

```python
#confusion matrix of dtree
confusion_matrix(y_pred1,y_test)#
```

```python
#confusion matrix of random forest classification
confusion_matrix(y_pred2,y_test)
```

we can see that the confusion matrix of random forest classification has more zeroes on non diagonal positions compared to confusion matrix of dtree.
therefore, random forest classification is better.
